```python
# fraction of filaments that might sever due to bending.

kT = 4.114 # pN nm

# based on De La Cruz BJ 2015: elastic energy associated with filament fragmentation.
#  On the one hand, De La CRuz incorporates twisting and twist-bend coupling, which will increase the stored elastic energy. 
# SO for a given bending and length, filaments will store more energy and fragment with less bending and for longer filaments.
# On the other hand, the estimate in de la cruz 2015 is for a filament shose ends are constrined in terms of their orientation, so theywill sever at a lower bending energy.
# See the final figure of their paper. So we treat this value as a lower limit, since one end is free in our simulations, and neither end bends significantly at the ends. 

severing_energy = 28*kT

print('severing energy = '+str(severing_energy))

# use pivot table to identify number of columns in the ends dataframe. other runs, timepoints will be 0
branched_actin_bound_ends_bending['severed']=branched_actin_bound_ends_bending['bending_energy']>severing_energy

data_frame_time_count_sever = branched_actin_bound_ends_bending.pivot_table(index=['run', 'time'], values='severed',  aggfunc='sum')
# .groupby(['run', 'time']).count()

data_frame_time_count_sever_mean = data_frame_time_count_sever.groupby('time').mean().severed
data_frame_time_count_sever_std  = data_frame_time_count_sever.groupby('time').std().severed

plt.figure(figsize=[5,4])
plot_multiple_errorbars(data_frame_time_count_sever_mean,data_frame_time_count_sever_std, 'blue', 'Severed filaments')

plt.savefig('../figures/'+pref+'_severed_filaments_time.png')
plt.savefig('../figures/'+pref+'_severed_filaments_time.pdf')

# number of severed filaments (all times)
mean_severed_filaments = data_frame_time_count_sever.mean()
std_severed_filaments  = data_frame_time_count_sever.std()

print('Number of severed filaments for energy over '+str(severing_energy)+' pN nm: '+str(mean_severed_filaments)+ ' ± ' +str(std_severed_filaments))

# data_frame_time_count_std

```
